### Comments based on Team Sync:
* to better capture the response rate for each column, we are defining "NaN" to be any code (e.g, 98, 99) or empty string that consitutes a non-acceptable response. May consider adding "IDK" to this bucket at a later time.
#### Game Plan:
* replace empty strings and codes 98, 99 (etc) with "NaN"
* obtain a ratio (NaN/column count) to obtain response rate for each column

```python
#please scroll down to A10: Which of the following best describes your [spouse's/partner's] current employment or work status?
for (columnName, columnData) in skinny_df.iteritems(): 
    print(columnName)
    print(skinny_df[columnName].value_counts())
    print()
```

#### Isolating first case of "non-acceptable" response as defined above:

```python
(skinny_df['A10'] == ' ').sum()
```

```python
response_rate = float((skinny_df['A10'] == ' ').sum()/skinny_df['A10'].count())
response_rate_percent = round(response_rate*100, 2)
print('response rate: ' + str(response_rate_percent)+'%')
```

#### Another interesting case appears with C5_2012

C5_2012: Do you [or your spouse/partner] regularly contribute to a retirement account like a [Thrift Savings Plan (TSP),] 401(k) or IRA? [2012 base]

```python
response_rate = float(((skinny_df['C5_2012'] == '99').sum()+(skinny_df['C5_2012'] == ' ').sum())/skinny_df['C5_2012'].count())
response_rate_percent = round(response_rate*100, 2)
print('response rate: ' + str(response_rate_percent)+'%')
```

#### Attempt at converting blanks to NA

```python
skinny_df_copy = skinny_df.copy()
```

```python
skinny_df_copy = skinny_df_copy.replace(r'^\s*$', np.NaN, regex=True)
```
